```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load image
img = Image.open(r"C:\Users\varad\OneDrive\Desktop\Sameer_19.jpg").convert("RGB")

# Preprocessing
preprocess = transforms.Compose([
    transforms.Resize(520),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])
input_tensor = preprocess(img).unsqueeze(0)  # Add batch dimension

# Load model
model = models.deeplabv3_resnet101(pretrained=True)
model.eval()  # 🔴 This is crucial

# Inference
with torch.no_grad():
    output = model(input_tensor)['out'][0]
    pred = output.argmax(0)  # Class prediction per pixel

# Visualize
plt.imshow(pred.numpy())
plt.axis('off')
plt.show()
```

```python
import numpy as np
import torchvision.transforms.functional as F

# Convert original image to tensor (unnormalized, same size as pred)
resized_img = transforms.Resize(520)(img)
img_tensor = F.to_tensor(resized_img)

# 1. Get background mask where predicted class == 0
background_mask = (pred == 0).numpy()  # shape: (H, W)

# 2. Expand mask to match color channels
background_mask_3ch = np.stack([background_mask]*3, axis=0)  # shape: (3, H, W)

# 3. Apply mask to image
background_only = img_tensor.numpy() * background_mask_3ch

# 4. Display result
plt.imshow(np.transpose(background_only, (1, 2, 0)))  # Convert to HWC
plt.axis('off')
plt.title("Only Background Pixels")
plt.show()
```

```python
import cv2
import numpy as np
from PIL import Image

# Convert PIL image to OpenCV format (for color space conversion)
img_cv = np.array(resized_img)[:, :, ::-1]  # Convert RGB to BGR

# 1. Get background mask
bg_mask = (pred == 0).numpy().astype(np.uint8)  # shape: (H, W)

# 2. Apply mask to image
masked_bg = cv2.bitwise_and(img_cv, img_cv, mask=bg_mask)

# 3. Convert to HSV
hsv_img = cv2.cvtColor(masked_bg, cv2.COLOR_BGR2HSV)

# 4. Only consider pixels where bg_mask == 1
h_values = hsv_img[:, :, 0][bg_mask == 1]
s_values = hsv_img[:, :, 1][bg_mask == 1]
v_values = hsv_img[:, :, 2][bg_mask == 1]

# 5. Compute standard deviations
std_h = np.std(h_values)
std_s = np.std(s_values)
std_v = np.std(v_values)

# 6. Heuristic: background is uniform if all stds < threshold
threshold = 10  # tweakable
is_uniform = (std_h < threshold) and (std_s < threshold) and (std_v < threshold)

# 7. Print result
print(f"Std Devs — H: {std_h:.2f}, S: {std_s:.2f}, V: {std_v:.2f}")
print("🟩 Uniform background detected!" if is_uniform else "🟥 Background is not uniform.")
```
