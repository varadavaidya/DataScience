```python
df = (data[['Ice Cream Sales','Temperature']]).head(100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
plt.scatter(df['Ice Cream Sales'],df['Temperature'])
plt.xlabel('Temperature (C)')
plt.ylabel('Ice Cream Sales')
plt.title('Ice cream sales vs Temperature')
plt.grid(True)
```

```python
x_train= df['Ice Cream Sales']
```

```python
y_train=df['Temperature']
```

```python
(x_train[0],y_train[0])
```

```python
(x_train[-1:],y_train[-1:])
```

```python
#assume 
w=100
b=50
```

```python
def compute_model_output(x,w,b):
    m=len(x_train)
    f_wb=[0]*m
    for i in range(m):
        f_wb[i]=int((w*x_train[i]+b)/100)
    return f_wb 
```

```python
temp_f_wb=compute_model_output(x_train,w,b)
```

```python
print(temp_f_wb)
```

```python
plt.plot(x_train,temp_f_wb,c='b',label='model_prediction')
plt.scatter(x_train,y_train,marker='x',c='r',label='ActualValues')
plt.xlabel('sales')
plt.ylabel('temp')
plt.title('Ice Cream Sales')
plt.legend()
```

```python
w=100
b=0
```

```python
pred=compute_model_output(x_train,w,b)
```

```python
plt.scatter(x_train,pred,marker='*',c='r',label='actualValues')
plt.plot(x_train,pred,c='b',label='predictions')
plt.xlabel('sales')
plt.ylabel('temp')
plt.title('pred when when intercept is 0')
plt.legend()
```
