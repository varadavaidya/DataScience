<p style="font-size:28px;">Tensorflow sequential model with compile function</p>

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras import Sequential 
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

```python
x=np.array([[200.0,17.0],
           [120.0,5.0],
           [425.0,20.],
           [212.0,18.0]])
```

```python
y=np.array([1,0,0,1])
```

```python
x= (x - np.mean(x, axis=0)) / np.std(x, axis=0)
```

```python
x
```

```python
layer_1=Dense(units=2,name='L1',activation='sigmoid')
```

```python
a1=layer_1(x)
```

```python
layer_2=Dense(units=1,name='L2',activation='sigmoid')
```

```python
a2=layer_2(a1)
```

```python
# tensor flow assumes parameters

model=Sequential([layer_1,layer_2])
```

```python
model.summary()
```

```python
model.get_layer('L1').get_weights()
```

```python
model.get_layer('L2').get_weights()
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01)
)
```

```python
model.fit(x,y,epochs=1000)
```

```python
model.get_weights()
```

```python
model.predict(np.array([[-0.34758084,  0.34050261]]))
```

## forward propagation   

```python
def sigmoid(z):
    return (1/(1+np.exp(-z)))
```

```python
def dense(a_in,w,b):
    units=w.shape[1]
    a_out=np.zeros(units)
    for i in range(units):
        z=np.dot(w[:,i],a_in)+b[i]
        a_out[i]=sigmoid(z)
    return a_out
```

```python
def sequential(x,w1,b1,w2,b2):
    a_1=dense(x,w1,b1)
    a_2=dense(a_1,w2,b2)
    return a_2
```

```python
def predict(x,w1,b1,w2,b2):
    res=sequential(x,w1,b1,w2,b2)
    if res>=0.5:
        return 1
    else:
        return 0
```

<p style="font-size:30px;">Using Vectorization</p>

```python
def dense_vectorized(x,W,b):
    res= np.matmul(x,W)+b
    return sigmoid(-res)
```

```python
def sequential_vectorization(x,W1,b1,W2,b2):
    a1=dense_vectorized(x,W1,b1)
    a2=dense_vectorized(a1,W2,b2)
    return a2
```

```python
sequential_vectorization(x[0],w1,b1,w2,b2)
```

```python
sequential_vectorization(x[1],w1,b1,w2,b2)
```

```python
sequential_vectorization(x[2],w1,b1,w2,b2)
```
