```python
def compute_parameters(x):
    m,n=x.shape
    mu=np.zeros(n)
    var=np.zeros(n)
    for i in range(n):
        mu[i]=float(sum(x[:,i]))/m
        var[i]=float(sum((x[:,i]-mu[i])**2))/m
    return mu,var
        
```

```python

def select_threshold(y_val, p_val):
    best_epsilon = 0
    best_F1 = 0
    F1 = 0
    step_size = (max(p_val) - min(p_val)) / 1000
    for epsilon in np.arange(min(p_val), max(p_val), step_size): 
        tp=0
        fp=0
        fn=0
        for i in range(y_val.shape[0]):
            if p_val[i]<epsilon and y_val[i]==1:
                tp+=1
            elif p_val[i]<epsilon and y_val[i]==0:
                fp+=1
            elif p_val[i]>epsilon and y_val[i]==1:
                fn+=1
        if tp + fp == 0 or tp + fn == 0:
            continue
            
        prec=tp/(tp+fp)
        rec=tp/(tp+fn)
        
        F1=(2*prec*rec)/(rec+prec)

        if F1 > best_F1:
            best_F1 = F1
            best_epsilon = epsilon
        
    return best_epsilon, best_F1
```

```python
def p_normal(x,mu,var):
    z=(x-mu)/(var**2)
    d=math.sqrt(2*math.pi)*var
    n=np.exp(-(z**2))
    return n/d
```

```python
def predictions(x_val,mu,var):
    m,n=x_val.shape
    pred=np.ones(m)
    for i in range(m):
        for j in range(n):
            pred[i]*=p_normal(x_val[i][j],mu[j],var[j])
    return pred
```
